```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('data/data_datetime.csv', sep=',')
df_train.info()
```

```python
encoder = preprocessing.LabelEncoder()
df_train['TAG'] = encoder.fit_transform(df_train['TAG'])
df_train['TAG'] = df_train['TAG'].astype('int')
```

```python
fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = (20, 20)
violin = ax.violinplot([df_train['MELT_TEMP'], df_train['MOTORSPEED'],df_train['MELT_WEIGHT'], df_train['INSP']], showmeans=True)
# ax.set_ylim(0, 10000)
ax.set_xticks(np.arange(1, 5))
ax.set_xticklabels(['MELT_TEMP', 'MOTORSPEED', 'MELT_WEIGHT', 'INSP'])
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```
